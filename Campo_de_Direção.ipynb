```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def campo_direcao(f, xlim=(-5, 5), ylim=(-5, 5), densidade=20, titulo="Campo de Direo"):
    """
    A funo plota o campo de direo de uma EDO do tipo dy/dx = f(x, y)

    Parmetros:
    - f: funo f(x, y)
    - xlim: tupla (xmin, xmax)
    - ylim: tupla (ymin, ymax)
    - densidade: nmero de vetores no eixo x e y
    - titulo: ttulo do grfico
    """
    # O Linspace  uma funo especfica do NumPy que gera nmeros espaados de forma uniforme em um intervalo especificado. Ele retorna um arranjo de valores igualmente espaados dentro de um intervalo definido.

    x = np.linspace(xlim[0], xlim[1], densidade)
    y = np.linspace(ylim[0], ylim[1], densidade)
    X, Y = np.meshgrid(x, y) # Gera duas matrizes X e Y contendo as coordenadas de cada ponto na grade 2D.

    U = np.ones_like(X)            # dx = 1
    V = f(X, Y)                    # dy = f(x, y)

    # Normalizao
    magnitude = np.sqrt(U**2 + V**2)
    U_norm = U / magnitude
    V_norm = V / magnitude

    plt.figure(figsize=(10, 6))
    plt.quiver(X, Y, U_norm, V_norm, angles='xy')
    plt.title(titulo, fontsize=14)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.show()

# Campo 1: f(x, y) = 3xy
campo_direcao(lambda x, y: 3 * x * y, xlim=(-3, 3), ylim=(-3, 3), titulo=r"$\frac{dy}{dx} = 3xy$")

# Campo 2: f(x, y) = y(2y - 1)
campo_direcao(lambda x, y: y * (2 * y - 1), xlim=(-3, 3), ylim=(-2, 2), titulo=r"$\frac{dy}{dx} = y(2y - 1)$")
```
