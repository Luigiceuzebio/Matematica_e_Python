```python
import numpy as np
import matplotlib.pyplot as plt

def calcula_probabilidade(n=10000):
    X = np.random.uniform(-1, 1, n) #Funo para gerar as variveis aleatrias entre [-1, 1]
    Y = np.random.uniform(-1, 1, n) #Funo para gerar as variveis aleatrias entre [-1, 1]
    
    I = (X**2 + Y**2 <= 1).astype(int) 
    
    probabilidade = np.sum(I) / n #Funo de somatrio da biblioteca Numpy
    return probabilidade, X, Y, I

n = 10000

probabilidade_aproximada, X, Y, I = calcula_probabilidade(n)
print(f'Probabilidade aproximada com n={n}: {probabilidade_aproximada}')

probabilidade_acumulada = np.cumsum(I) / np.arange(1, n+1) #Dividimos a soma acumulada dos indicadores pelo nmero de pontos considerados at aquele ponto

plt.figure(figsize=(9, 5))
plt.plot(range(1, n+1), probabilidade_acumulada, label='Probabilidade acumulada')
plt.axhline(y=np.pi/4, color='r', linestyle='--', label='Valor Esperado: /4')
plt.xlabel('Nmero de pontos (n)')
plt.ylabel('Probabilidade acumulada')
plt.title('Convergncia da probabilidade acumulada para /4')
plt.legend()
plt.grid(True)
plt.show()
```
